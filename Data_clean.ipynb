```python
cols=['monthly_rent','credit_score','emergency_fund']

for col in cols:
    data[col]=data[col].fillna(data[col].median())
```

```python
# convert to numeric, coercing  error to nan
data['bank_balance']=pd.to_numeric(data['bank_balance'],errors='coerce')
median_balance = data['bank_balance'].median()
data['bank_balance_1'] = data['bank_balance'].fillna(median_balance)
```

```python
data['age']=pd.to_numeric(data['age'],errors='coerce')
data['age']=data['age'].fillna(0).astype(int)
median_age=data.loc[data['age']!=0,'age'].median()
data['age']=data['age'].replace(0,median_age)
```
